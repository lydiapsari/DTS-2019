## 4. Analisis Model Regresi

Melakukan training dan testing data dengan nilai random 

```python
import numpy as np
msk = np.random.rand(len(data1)) < 0.8
train = data1[msk]
test = data1[~msk]
```

#### Features Average Score (SAT Reading) dan Average Score (SAT Writing)
Menampilkan scatter plot dari data training untuk features Average Score (SAT Reading) dan Average Score (SAT Writing)

```python
plt.scatter(train[['Average_Score_(SAT_Reading)']], train[['Average_Score_(SAT_Writing)']],  color='red')
plt.xlabel("Average_Score_(SAT_Reading)")
plt.ylabel("Average_Score_(SAT_Writing)")
plt.show()
```

Mencari nilai b0 dan b1, dan menghitung nilai Mean absolute error, Residual sum of squares (MSE), dan R2-Score

```python
from sklearn import linear_model
from sklearn.metrics import r2_score

regr1 = linear_model.LinearRegression()
x1_train = train[['Average_Score_(SAT_Reading)']]
y1_train = train[['Average_Score_(SAT_Writing)']]
regr1.fit (x1_train, y1_train)

x1_test = test[['Average_Score_(SAT_Reading)']]
y1_test = test[['Average_Score_(SAT_Writing)']]
y1_test_ = regr1.predict(x1_test)

# The coefficients
print ('Coefficients: ', regr1.coef_)
print ('Intercept: ',regr1.intercept_)
print("Mean absolute error: %.2f" % np.mean(np.absolute(y1_test_ - y1_test)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y1_test_ - y1_test) ** 2))
print("R2-score: %.2f" % r2_score(y1_test_ , y1_test) )
```

Menampilkan scatter plot untuk pemodelan regresi linier

```python
plt.scatter(x1_train[['Average_Score_(SAT_Reading)']], y1_train[['Average_Score_(SAT_Writing)']],  color='red')
plt.plot(x1_train, regr1.coef_[0][0]*x1_train + regr1.intercept_[0], '-y', color='blue')
plt.xlabel("Average_Score_(SAT_Reading)")
plt.ylabel("Average_Score_(SAT_Writing)")
```

<b>Analisis:</b>
Berdasarkan nilai R2-Score yang didapatkan yaitu sebesar 0.98 dapat diambil kesimpulan bahwa pemodelan regresi linier ini cocok dengan nilai R2-Score yang mendekati nilai 1. 
Selain itu juga dapat diambil kesimpulan bahwa nilai Average Score (SAT Reading) dan Average Score (SAT Writing) berbanding lurus dan menandakan bahwa apabila nilai Average Score (SAT Reading) tinggi, maka nilai Average Score (SAT Writing) juga akan tinggi karena saling berpengaruh.

#### Features Average Score (SAT Math) dan Average Score (SAT Reading)

## 4. Analisa Clustering

#### Features Average Score (SAT Reading) dan Average Score (SAT Writing)
Menentukan jumlah cluster dengan menggunakan Metode Elbow untuk features Average Score (SAT Reading) dan Average Score (SAT Writing)

```python
X = data1.iloc[:, [19, 20]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Metode Elbow')
plt.xlabel('Jumlah clusters')
plt.ylabel('WCSS')
plt.show()
```

Berdasarkan grafik diatas dapat ditentukan jumlah cluster yang dihasilkan adalah 2

```python
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
labels = kmeans.labels_
print(labels)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 20, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 20, c = 'blue', label = 'Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', marker='*', label = 'Centroids')
plt.title('Analisa Average Score (SAT Reading) dan Average Score (SAT Writing)')
plt.xlabel('Average Score (SAT Reading)')
plt.ylabel('Average Score (SAT Writing)')
plt.legend()
plt.legend()
plt.show()
```

Pada grafik diatas dapat dilihat terdapat 3 cluster dengan warna yang berbeda. 
1. Cluster 1 = cluster 1 ditandai dengan warna merah dengan nilai rata-rata untuk SAT Math dan SAT reading rendah.
2. Cluster 2 = cluster 2 ditandai dengan warna biru dengan nilai rata-rata untuk SAT Math dan SAT reading tinggi.
4. Centroid = centroid ditandai dengan simbol bintang

Mencari nilai silhouette score

```python
from sklearn.metrics import silhouette_score

print(silhouette_score(X, kmeans.labels_))
```

Nilai silhouette score sebesar 0.6717292883368692 menunjukkan bahwa data diatas cocok menggunakan 2 cluster. Ditunjukkan dengan grafik Silhouette Score Elbow for KMeans Clusterin

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans(random_state=0)
visualizer = KElbowVisualizer(model, k=(2,6), metric='silhouette', timings=False)
visualizer.fit(X)    
visualizer.poof()  
```

Dari grafik diatas terlihat bahwa cluster terbaik untuk data adalah 2 cluster
